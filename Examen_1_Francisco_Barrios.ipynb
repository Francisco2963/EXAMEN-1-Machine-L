Realizaremos una funcion donde utilizaremos los cuartiles para encontrar los outliers.

como creamos la variable objetivo tipo (tinto o blaco) y es un object, la funcion no la tomara en cuenta devolviendo None, y recorrera los cuartiles buscando los outliers

```python
def calcular_outliers(df, columna):
  if df_final[columna].dtype == 'object':
    return None, None, None, None
  else:

   Q1 = df_final[columna].quantile(0.25)
   Q3 = df_final[columna].quantile(0.75)
   IQR = Q3 - Q1
   outliers = df_final[(df_final[columna] < Q1 - 1.5*IQR) | (df[columna] > Q3 + 1.5*IQR)]
   return outliers, Q1, Q3, IQR

outliers_fixed_acidity, Q1_fixed_acidity, Q3_fixed_acidity, IQR_fixed_acidity = calcular_outliers(df_final, 'fixed acidity')
outliers_volatile_acidity, Q1_volatile_acidity, Q3_volatile_acidity, IQR_volatile_acidity = calcular_outliers(df_final, 'volatile acidity')
outliers_citric_acid, Q1_citric_acid, Q3_citric_acid, IQR_citric_acid = calcular_outliers(df_final, 'citric acid')
outliers_residual_sugar, Q1_residual_sugar, Q3_residual_sugar, IQR_residual_sugar = calcular_outliers(df_final, 'residual sugar')
outliers_chlorides, Q1_chlorides, Q3_chlorides, IQR_chlorides = calcular_outliers(df_final, 'chlorides')
outliers_free_sulfur_dioxide, Q1_free_sulfur_dioxide, Q3_free_sulfur_dioxide, IQR_free_sulfur_dioxide = calcular_outliers(df_final, 'free sulfur dioxide')
outliers_total_sulfur_dioxide, Q1_total_sulfur_dioxide, Q3_total_sulfur_dioxide, IQR_total_sulfur_dioxide = calcular_outliers(df_final, 'total sulfur dioxide')
outliers_density, Q1_density, Q3_density, IQR_density = calcular_outliers(df_final, 'density')
outliers_pH, Q1_pH, Q3_pH, IQR_pH = calcular_outliers(df_final, 'pH')
outliers_sulphates, Q1_sulphates, Q3_sulphates, IQR_sulphates = calcular_outliers(df_final, 'sulphates')
outliers_alcohol, Q1_alcohol, Q3_alcohol, IQR_alcohol = calcular_outliers(df_final, 'alcohol')
outliers_quality, Q1_quality, Q3_quality, IQR_quality = calcular_outliers(df_final, 'quality')
outliers_tipo, Q1_tipo, Q3_tipo, IQR_tipo = calcular_outliers(df_final, 'tipo')


print(f'Outliers en fixed acidity: \n{len(outliers_fixed_acidity)}')

print(f'Outliers en volatile acidity: \n{len(outliers_volatile_acidity)}')

print(f'Outliers en citric acid: \n{len(outliers_citric_acid)}')

print(f'Outliers en residual sugar: \n{len(outliers_residual_sugar)}')

print(f'Outliers en chlorides: \n{len(outliers_chlorides)}')

print(f'Outliers en free sulfur dioxide: \n{len(outliers_free_sulfur_dioxide)}')

print(f'Outliers en total sulfur dioxide: \n{len(outliers_total_sulfur_dioxide)}')

print(f'Outliers en density: \n{len(outliers_density)}')

print(f'Outliers en pH: \n{len(outliers_pH)}')

print(f'Outliers en sulphates: \n{len(outliers_sulphates)}')

print(f'Outliers en alcohol: \n{len(outliers_alcohol)}')

print(f'Outliers en quality: \n{len(outliers_quality)}')
```

```python
import pandas as pd
import numpy as np

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.pipeline import make_pipeline
```

```python

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, roc_auc_score
import matplotlib.pyplot as plt
```

Ahora cambiaremos nuestra variable objetivo (target) de categorica a numerica

```python
y = df_final['tipo']
X = df_final.drop('tipo', axis = 'columns')
```

```python
X
```

```python
y
```

Con LabelEncoder cambiamos nuestra variable objetivo de categorica a numerica, por que elegimos labelEncoder y no OrdinalEncoder?

Porque necesitamos que la clasificacion del vino sea en una sola columna como 1 y 0. si elegimos OrdinalEncoder nos va a crear dos columnas con valores de 0 y 1 y no queremos separar nuestra variable objetivo en dos columnas

```python
label = LabelEncoder()
y = label.fit_transform(y)
```

```python
y
```

```python
X.info()
```

```python
# Dividimos los datos en conjunto de entrenamiemto y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#columnas numericas
variables_num =["fixed acidity","volatile acidity","citric acid","residual sugar","chlorides",
                "free sulfur dioxide","total sulfur dioxide","density","pH","sulphates",
                "alcohol","quality"]
num_transformer = Pipeline(
    steps=[
        ("imputer",SimpleImputer(strategy="mean")),
        ("scaler",StandardScaler())
    ]
)


#se incluye un columntransformer para agrupar las transformaciones realizadas
preprocessor = ColumnTransformer(
    transformers=[
        ("num",num_transformer,variables_num)
    ]
)

# Pipeline final que incluye imputación condicional numerica y modelo

pipeline_KNNC = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("model", KNeighborsClassifier(n_neighbors=20))
])

pipeline_LR = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("model", LogisticRegression(penalty='l2', C=1.0, solver='lbfgs',
                                 max_iter=100, tol=1e-4, random_state=42))
])



```

Con ColumnTransformer y Pipeline preparamos los datos para ser utilizados en los modelos de Regresion Logistica y KNN

```python
# Entrenar el modelo
pipeline_KNNC.fit(X_train, y_train)
pipeline_LR.fit(X_train, y_train)
```

```python
# Realizar Predicciones
y_pred_LR = pipeline_LR.predict(X_test)
y_pred_KNNC = pipeline_KNNC.predict(X_test)
```

# KNN

```python
cm=confusion_matrix(y_test,y_pred_KNNC)
print("Reporte de Clasificacion por KNNC")
print(classification_report(y_test,y_pred_KNNC))
cm2=ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=np.unique(y))
fig, ax=plt.subplots(figsize=(8,6))
cm2.plot(ax=ax)
plt.title('Matriz de confusion Clasificacion del Vino')
plt.ylabel('Valor Real')
plt.xlabel('Valor Predicho')
plt.show()
```

COMENTARIO

El modelo tiene una precision perfecta para el vino blanco (1.00) y para el tinto de (0.98).

el modelo se equivoca poco, casi nada, las muestras de vino blanco es de 1163 muestras, clasificando solo 9 muestras como tinto.

Para el vino tinto, las muestras son 433, clasificando como vino blanco 5 muestras.

El recall tiene un 99% para ambos tipos de vinos, lo que indica que clasifica bien las muestras comos e dijo una lineas arriba para el vino blaco clasifico 9 muestras como tinto y 5 muestras como blanco
